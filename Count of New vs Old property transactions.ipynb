```python
import pandas as pd
```

```python
df = pd.read_csv('eng_pp_9523_eda_pivoted.csv')
```

```python
# Count of New vs Old property transactions
new_property_count = df[df['Old/New'] == 'N'].shape[0]
old_property_count = df[df['Old/New'] == 'Y'].shape[0]

# Print result
print("Number of New Property Transactions:", new_property_count)
print("Number of Old Property Transactions:", old_property_count)
```

```python
# You can also calculate the percentage here
total_transactions = new_property_count + old_property_count
new_property_percentage = (new_property_count / total_transactions) * 100
old_property_percentage = (old_property_count / total_transactions) * 100

print("Percentage of New Property Transactions:", new_property_percentage, "%")
print("Percentage of Old Property Transactions:", old_property_percentage, "%")
```

```python
# Pie Chart Plot
import matplotlib.pyplot as plt

# Define data
df_oldnew = {
    "Property Type": ["New", "Old"],
    "Number of Transactions": [20284102, 2369884],
    "Percentage": [89.53877697284707, 10.461223027152926]
}

# Create colors list
colors = ['skyblue', 'blue']

# Create Pie chart
plt.figure(figsize=(8, 8))
plt.pie(df_oldnew['Number of Transactions'], labels=df_oldnew['Property Type'], autopct="%1.1f%%", startangle=90, colors=colors)
plt.title("Distribution of New vs. Old Property Transactions (1995-2023)")
plt.axis('equal')  # Equal aspect ratio ensures a circular pie chart

# Display plot
plt.show()
```
