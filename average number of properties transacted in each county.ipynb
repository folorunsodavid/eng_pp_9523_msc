```python
import pandas as pd
```

```python
df = pd.read_csv('eng_pp_9523_eda_pivoted.csv')
```

```python
# Calculate average number of properties sold in each county
average_sold_per_county = df.groupby('County')['Year'].count().div(len(df['Year'].unique()))

# Print the result
print(average_sold_per_county)
```

```python
# Calculate yearly number of properties sold in each county
yearly_sold_per_county = df.groupby(['County', 'Year']).size().unstack()

# Print the result
print(yearly_sold_per_county)
```

```python
yearly_sold_per_county
```

```python
yearly_sold_per_county.to_csv('Property Yearly Transactions by County (1995 - 2023).csv')
```

```python
import matplotlib.pyplot as plt

# Plot average number of properties sold in each county
plt.figure(figsize=(10, 6))  # Adjust the figure size as desired
plt.bar(average_sold_per_county.index, average_sold_per_county.values, width = 0.7)
plt.xlabel('County')
plt.ylabel('Average Number of Properties Sold')
plt.title('Average Number of Properties Sold in Each County (1995-2023)')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
