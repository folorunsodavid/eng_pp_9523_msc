```python
import pandas as pd
```

```python
df = pd.read_csv('eng_pp_9523_eda_pivoted.csv')
```

```python
df.head()
```

```python
# Calculate Average Property Types Transaction by County
average_transactions = df.groupby('County')[['Detached', 'Semi-detached', 'Terraced', 'Flat', 'Others']].mean()

# Print the result
print(average_transactions)
```

```python
# Calculate yearly number of properties sold in each county
yearly_sold_per_county = df.groupby(['County', 'Year']).size().unstack()

# Print the result
print(yearly_sold_per_county)
```
