```python
import pandas as pd
```

```python
df = pd.read_csv("eng_pp_9523.csv")
```

```python
# Specify the columns to drop
columns_to_drop = ['postcode', 'Transaction_unique_identifier','Duration', 'PAON', 'SAON', 'PPDCategory_Type', 'Record_Status - monthly_file_only']
```

```python
# Drop the specified columns
englandhousing_column_dropped = df.drop(columns=columns_to_drop)
```

```python
df = pd.DataFrame(englandhousing_column_dropped)
```

```python
df
```

```python
# Specify desired column order (adjust as needed)
df_reordered = ['Date_of_Transfer', 'Property_Type', 'price', 'Old/New', 'Street', 'Locality', 'Town/City', 'District', 'County']
```

```python
# Rearrange columns
df = df[df_reordered]
```

```python
df
```

```python
# Convert the date column to datetime
df['Date_of_Transfer'] = pd.to_datetime(df['Date_of_Transfer'])
```

```python
df2 = df.set_index('Date_of_Transfer', inplace=True)
```

```python
df2
```

```python
# Save the modified dataset
df2.to_csv('eng_pp_9523_drop_unused_columns.csv', index=True)
```
