```python
#Data cleaning - dropping null values
df2_cleaned = df2.dropna()
```

```python
# Convert the date column to datetime
df2['Date_of_Transfer'] = pd.to_datetime(df2['Date_of_Transfer'])
```

```python
# Extract year and month as separate columns
df2['Year'] = df2['Date_of_Transfer'].dt.year
df2['Month'] = df2['Date_of_Transfer'].dt.month_name()
```

```python
# Drop the original date column if needed
df2.drop('Date_of_Transfer', axis=1, inplace=True)
```

```python
# Print the resulting DataFrame
df2
```
