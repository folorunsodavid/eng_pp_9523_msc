```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("eng_pp_9523_eda_pivoted.csv")
```

```python
# Group data by county and calculate the proportion of each property type
county_prop_types = df.groupby('County')[['Detached', 'Semi-detached', 'Terraced', 'Flat', 'Others']].sum()
county_prop_types = county_prop_types.div(county_prop_types.sum(axis=1), axis=0) * 100
```

```python
# Analyze the results
print(county_prop_types.describe())
```

```python
# Create the bar chart with wider bars
bar_width = 0.7  # Adjust the value to your desired width
ax = county_prop_types.plot(kind='bar', stacked=True, figsize=(10, 6), width=bar_width)

# Set vertical x-axis labels
plt.xticks(rotation=90)  # Rotate labels by 90 degrees
plt.xlabel('County')
plt.ylabel('Proportion (%)')
plt.title('Proportion of Property Types by County')
plt.subplots_adjust(bottom=0.3)  # Adjust space for x-axis labels
```

```python
#create a heatmap
plt.pcolor(county_prop_types)
plt.colorbar(label='Proportion (%)')
plt.xlabel('County')
plt.ylabel('Property Type')
plt.xticks(rotation=0)
```
